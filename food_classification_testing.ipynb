```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
#load our model
cnn = tf.keras.models.load_model('/content/trained_model (1).h5')
```

```python
#performing prediction on single image
```

```python
import cv2
image_path= '/content/drive/MyDrive/test/banana/Image_10.jpg'
img= cv2.imread(image_path)
plt.imshow(img)
plt.title("TEST IMAGE")
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
#testing model
```

```python
image= tf.keras.preprocessing.image.load_img(image_path, target_size= (64,64))
input_arr= tf.keras.preprocessing.image.img_to_array(image)
input_arr= np.array([input_arr]) # converting single image to batch
predictions= cnn.predict(input_arr)
```

```python
print(predictions)
```

```python
test_set= tf.keras.utils.image_dataset_from_directory('/content/drive/MyDrive/test',
labels='inferred',
label_mode='categorical',
class_names=None,
color_mode='rgb',
batch_size=32,
image_size=(64, 64),
shuffle=True,
seed=None,
validation_split=None,
subset=None,
interpolation='bilinear',
follow_links=False,
crop_to_aspect_ratio=False
                                       )
```

```python
test_set.class_names
```

```python
result_index= np.where(predictions[0]== max(predictions[0]))
print(result_index[0][0])
```
